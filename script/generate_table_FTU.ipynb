```python
import pandas as pd
```

```python
vasculature=pd.read_csv('Vasculature Expanded ASCT+B Table - DataTable.csv')
df = pd.DataFrame(columns = vasculature.columns)
vasculature.head()
```

```python
end_vessel = ['glomerular capillary','vasa recta of kidney','subsegmental pulmonary artery','short branch of vasa recta of colon','spleen central arteriole','right hepatic artery','popliteal artery','thymic artery','superior left bronchial artery','dorsal pancreatic artery']
```

```python
for z in end_vessel:
    for i in vasculature.index:
        if vasculature['Vessel'].loc[i]==z:
            pathfh=vasculature['PathFromHeart'].loc[i]
            pathfh = pathfh.split(';')
            print(pathfh)
            for a in pathfh:
                for b in vasculature.index:
                    if vasculature.Vessel.loc[b]==a:
                        df.loc[len(df.index)] = vasculature.loc[b]
```

```python
print(df)
```

```python
df.drop_duplicates(subset ="Vessel", ignore_index=True, keep = False, inplace = True)
print(df)
```

```python
df.to_csv('ftu.network_graph.csv')
```
